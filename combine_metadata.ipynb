```python
import pandas as pd
```

```python
annotations = pd.read_csv('stanford-cars/car_annotations.csv')
annotations.head()
```

```python
classes = pd.read_csv('stanford-cars/class_names2.csv', index_col = 0)
classes.head()
```

```python
full = annotations.merge(classes, how = 'left', left_on = 'class', right_on = 'class_index', validate = 'm:1')
```

```python
full.head()
```

```python
full.to_csv('stanford-cars/full_annotations.csv')
```
